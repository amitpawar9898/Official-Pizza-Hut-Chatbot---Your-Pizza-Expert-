```python
import gradio as gr
import random

def pizza_hut_bot(message, history):
    message_lower = message.lower()
    pizza_image_urls = [
        "/content/pizza-5179939_1280.jpg",
    ]
    if "burger" in message_lower or "hamburger" in message_lower or "cheeseburger" in message_lower:
        return "I apologize üòî, but Pizza Hut specializes exclusively in delicious pizzas and pizza-related items. How about we discuss our fantastic pizza menu instead?"
    elif "pizza" in message_lower or "topping" in message_lower or "crust" in message_lower or "order" in message_lower:
        pizza_info = "That's a great question about pizza! üçï At Pizza Hut, we have a wide variety of pizzas to satisfy every craving. You could try our classic Pepperoni, a cheesy Veggie, the meat-packed Meat Lovers, the tropical Hawaiian, or even our spicy Buffalo Chicken pizza! What sounds good to you? üòã"
        selected_image_url = random.choice(pizza_image_urls)
        return f"{pizza_info}\n\n![Delicious Pizza]({selected_image_url})"
    elif "delivery" in message_lower or "deliver" in message_lower or "driver" in message_lower or "delivery boy" in message_lower:
        return "Our dedicated delivery boy, Omkar, is always ready to bring your delicious pizza right to your doorstep! You can contact him at üìû 7498620252. What would you like to order today? üõµüçï"
    elif "deals" in message_lower or "promotions" in message_lower or "offers" in message_lower:
        return "We always have amazing deals at Pizza Hut! ‚ú® Right now, you can get our Buy One Get One Free offer on large pizzas, or check out our combo meals for great value! Which offer sounds good to you?"
    else:
        return "Welcome to Pizza Hut! üçï We're all about crafting the perfect pizza for you. What kind of pizza are you in the mood for today, or perhaps you'd like to hear about our latest deals? ‚ú®"

custom_css = """
body {
    background-color: lightgray; /* Changed to a solid color */
    background-size: cover;
    background-repeat: no-repeat;
    background-attachment: fixed;
    background-position: center center;
}
.gradio-container {
    background-color: rgba(255, 255, 255, 0.8); /* Semi-transparent white to make content readable */
    border-radius: 10px;
    padding: 20px;
}
h1.gradio-title {
    color: darkred; /* Changed to dark red for the title */
}
"""
demo = gr.ChatInterface(
    fn=pizza_hut_bot,
    title="Official Pizza Hut Chatbot - Your Pizza Expert!",
    description="Ask me anything about Pizza Hut's delicious pizzas and more! (P.S. We only talk about pizza!)",
    chatbot=gr.Chatbot(type='messages', allow_tags=True),
    type='messages',
    theme='huggingface',
    css=custom_css
)

demo.launch()
```
